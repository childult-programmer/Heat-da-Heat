```python
import json
from pathlib import Path
import random
import pprint
import pickle

from collections import defaultdict
```

```python
path = Path('../../json/all.json')

with open(path, 'r') as f:
    json_data = json.load(f)
```

```python
facility_date = defaultdict(int)

for _, _, _, p in json_data:
    facility_date[p.split('/', maxsplit=6)[-1].rsplit('(', maxsplit=2)[0]] += 1
```

```python
len_facility_date = defaultdict(list)

for k, v in facility_date.items():
    sep = k.rsplit('/', maxsplit=1)
    if sep[1].startswith('0'):
        len_facility_date[sep[0]].append(sep[1] + ':' + str(v))
    else:
        len_facility_date[k].append(str(v))
```
